## Bank Application [ Python project ]

```python
from tkinter import *
import os

master=Tk()
master.title('Bank App')
master.geometry('800x600')
##################################################################################################
def Deposite():
    global depo_screen
    depo_screen=Toplevel(master)
    depo_screen.title('Deposite')
    
    global depo_amount
    depo_amount=StringVar()
    
    file=open(log_ses_name,'r')
    dep_data=file.read().split('\n')
    dep_bal=dep_data[4].split(':')[1]
    
    Label(depo_screen,text='Deposit',font=('calibri',20,'bold'),fg='black',bg='green').grid(row=0,column=2)
    Label(depo_screen,text='Your current balance: Rs.'+dep_bal).grid(row=2,column=2)
    Label(depo_screen,text='Enter Amount: ').grid(row=3,column=1)
    
    Entry(depo_screen,textvariable=depo_amount).grid(row=3,column=2)
    
    Button(depo_screen,text='Deposite',command=Depo_finish,width=15).grid(row=4,column=2)
    Button(depo_screen,text='Close',command=depo_screen.destroy,width=15).grid(row=5,column=2)
    
    global depo_notif
    depo_notif=Label(depo_screen)
    depo_notif.grid(row=6,column=2)
    
###################################################################################################
def Depo_finish():
    
    dep_amount=depo_amount.get()
    depo_name=log_ses_name
    
    if dep_amount=='':
        depo_notif.config(text='*Amount is required.',fg='red')
        return
    if float(dep_amount)<=0:
        depo_notif.config(text="*Negitive amount can't be inserted.",fg='red')
        return
    
    file=open(depo_name,'r+')
    file_data=file.read()
    detail=file_data.split('\n')
    cur_bal=detail[4].split(':')[1]
    up_bal=float(cur_bal)+float(dep_amount)
    file_data=file_data.replace(cur_bal,str(up_bal))
    file.seek(0)
    file.truncate(0)
    file.write(file_data)
    file.close()
    
    Label(depo_screen,text='Your current balance: Rs.'+str(up_bal),fg='green').grid(row=2,column=2)
    
    depo_notif.config(text='Balance updated.',fg='green')
    
    
###################################################################################################
def Withdraw():
    
    global with_screen
    with_screen=Toplevel(master)
    with_screen.title('Withdraw')
    
    global with_amount
    with_amount=StringVar()
    
    file=open(log_ses_name,'r')
    with_data=file.read().split('\n')
    with_bal=with_data[4].split(':')[1]
    
    Label(with_screen,text='Withdraw',font=('calibri',20,'bold'),fg='black',bg='green').grid(row=0,column=2)
    Label(with_screen,text='Your current balance: Rs.'+with_bal).grid(row=2,column=2)
    Label(with_screen,text='Enter Amount: ').grid(row=3,column=1)
    
    Entry(with_screen,textvariable=with_amount).grid(row=3,column=2)
    
    Button(with_screen,text='Withdraw',command=With_finish,width=15).grid(row=4,column=2)
    Button(with_screen,text='Close',command=with_screen.destroy,width=15).grid(row=5,column=2)
    
    global with_notif
    with_notif=Label(with_screen)
    with_notif.grid(row=6,column=2)   
    
##################################################################################################
def With_finish():
    
    global wit_amount
    wit_amount=with_amount.get()
    
    global wit_name
    wit_name=log_ses_name
    
    if wit_amount=='':
        with_notif.config(text="*Amount is required.",fg='red')
        return
        
    if float(wit_amount)<=0:
        with_notif.config(text="*Negitive amount can't be inserted.",fg='red')
        return
    
    file=open(wit_name,'r+')
    file_data=file.read()
    detail=file_data.split('\n')
    cur_bal=detail[4].split(':')[1]
    
    if float(wit_amount)>float(cur_bal):
        with_notif.config(text="*Insufficient Balance.",fg='red')
        return
    
    up_bal=float(cur_bal)-float(wit_amount)
    file_data=file_data.replace(cur_bal,str(up_bal))
    file.seek(0)
    file.truncate(0)
    file.write(file_data)
    file.close()
    
    Label(with_screen,text='Your current balance: Rs.'+str(up_bal),fg='green').grid(row=2,column=2)
    
    with_notif.config(text='Balance updated.',fg='green')
    
    
        
        
##################################################################################################
def Finish_reg():
    name=reg_name.get()
    age=reg_age.get()
    gender=reg_gender.get()
    password=reg_password.get()
    balance=reg_balance.get()
    robotvalue=reg_robotvalue.get()
    all_account=os.listdir()
    
    if name=='' or age=='' or gender=='' or password=='' or balance==''or robotvalue==0 or name.isalpha()==False:
        reg_notif.config(text='*Sorry, All fields are required ! ',fg='red')
        return
    
    for reg_exist_name in all_account:
        if name==reg_exist_name:
            reg_notif.config(text='*Sorry, Account already exist ! ',fg='red')
        else:
            user_file=open(name,'w')
            user_file.write(f'Name:{name}\n')
            user_file.write(f'Age:{age}\n')
            user_file.write(f'Gender:{gender}\n')
            user_file.write(f'Password:{password}\n')
            user_file.write(f'Balance:{balance}\n')
            user_file.write(f'Robot_Val:{robotvalue}')
            user_file.close()
            
            con_screen=Toplevel(master)
            con_screen.title('Register')
            
            reg_screen.destroy()
            
            Label(con_screen,text='Congratulations! Your account has been created successfuly.',fg='green',font=('calibri',20,'bold')).grid(row=2,column=4)
            Button(con_screen,text='Back to home',command=con_screen.destroy,width=25).grid(row=4,column=4)
            break
            
            
            
            
##################################################################################################    
def Register(): 
    
    global reg_screen
    reg_screen=Toplevel(master)
    reg_screen.title('Register')
    
    global reg_name
    global reg_age
    global reg_gender
    global reg_password
    global reg_balance
    global reg_robotvalue
    
    reg_name=StringVar()
    reg_age=StringVar()
    reg_gender=StringVar()
    reg_password=StringVar()
    reg_balance=StringVar()
    reg_robotvalue=IntVar( )
        
        
    Label(reg_screen,text='Registration',font=('calibri',20,'bold'),fg='black',bg='green').grid(row=0,column=1)
    Label(reg_screen,text='Name:- ').grid(row=1,column=0)
    Label(reg_screen,text='Age:- ').grid(row=2,column=0)
    Label(reg_screen,text='Gender:- ').grid(row=3,column=0)
    Label(reg_screen,text='password:- ').grid(row=4,column=0)
    Label(reg_screen,text='Balance:- ').grid(row=5,column=0)
    
    Entry(reg_screen,textvariable=reg_name).grid(row=1,column=1)
    Entry(reg_screen,textvariable=reg_age).grid(row=2,column=1)
    Entry(reg_screen,textvariable=reg_gender).grid(row=3,column=1)
    Entry(reg_screen,textvariable=reg_password).grid(row=4,column=1)
    Entry(reg_screen,textvariable=reg_balance).grid(row=5,column=1)
    
    Checkbutton(reg_screen,text="I'm not robot.",variable=reg_robotvalue).grid(row=6,column=1)
    
    Button(reg_screen,text='Register',command=Finish_reg,width=15).grid(row=7,column=1)
    Button(reg_screen,text='Close',command=reg_screen.destroy,width=15).grid(row=8,column=1)
    
    global reg_notif
    reg_notif=Label(reg_screen)
    reg_notif.grid(row=9,column=1)
##################################################################################################

def Personal_detail():
    detail_screen=Toplevel(master)
    detail_screen.title('Personal Details')
    d_name=log_ses_name
    Label(detail_screen,text=d_name+"'s Personal Details",font=('calibri',20,'bold'),fg='black',bg='green').grid(row=0,column=1)
    
    detail_file=open(d_name,'r')
    file_data=detail_file.read().split('\n')
    detail_file.close()
    
    det_name=file_data[0]
    det_age=file_data[1]
    det_gender=file_data[2]
    det_password=file_data[3]
    det_balance=file_data[4]
    det_robot=file_data[5]
    
    Label(detail_screen,text=det_name).grid(row=1,column=1)
    Label(detail_screen,text=det_age).grid(row=2,column=1)
    Label(detail_screen,text=det_gender).grid(row=3,column=1)
    Label(detail_screen,text=det_password).grid(row=4,column=1)
    Label(detail_screen,text=det_balance).grid(row=5,column=1)
    Label(detail_screen,text=det_robot).grid(row=6,column=1)
    
    Button(detail_screen,text='Close',command=detail_screen.destroy,width=15).grid(row=7,column=1)
###################################################################################################
def Login_session():
    global log_ses_name
    global log_ses_password
    
    log_ses_name=log_name.get()
    log_ses_password=log_password.get()
    all_account=os.listdir()

    for log_exist_name in all_account:
        if log_exist_name==log_ses_name:
            file=open(log_ses_name,'r')
            file_data=file.read().split('\n') #now it will in object form-->now after split data into split.
            file_password=file_data[3].split(':')[1]
            file.close()#####################################################
            
            if file_password==log_ses_password:
                log_screen.destroy()
                             
                # after closing login page user's dashboard will display.
                dash_screen=Toplevel(master)
                dash_screen.title('Dashboard')
                
                Label(dash_screen,text="User's Dashboard",font=('calibri',20,'bold'),fg='black',bg='green').grid(row=0,column=1)
                Label(dash_screen,text='Welcome to your dashboard '+ log_ses_name).grid(row=2,column=1)
                
                Button(dash_screen,text='Personal Detail',command=Personal_detail,width=30).grid(row=3,column=1)
                Button(dash_screen,text='Deposite',command=Deposite,width=30).grid(row=4,column=1)
                Button(dash_screen,text='Withdraw',command=Withdraw,width=30).grid(row=5,column=1)
                Button(dash_screen,text='Log Out',command=dash_screen.destroy,width=15).grid(row=6,column=1)
                             
            else:
                log_notif.config(text='Password Incorrect',fg='red')
                break
        else:                     
            log_notif.config(text='No Account Found',fg='red') 
    
                             
###################################################################################################    
def Login():
    global log_screen                         
    log_screen=Toplevel(master)
    log_screen.title('Login')
                             
    global log_name
    global log_password
    
    log_name=StringVar()
    log_password=StringVar()
                             
    Label(log_screen,text='Login to your account',font=('calibri',20,'bold'),fg='black',bg='green').grid(row=0,column=1)
    Label(log_screen,text='Username:- ').grid(row=1,column=0)
    Label(log_screen,text='Password:- ').grid(row=2,column=0)
                         
    Entry(log_screen,textvariable=log_name).grid(row=1,column=1)
    Entry(log_screen,textvariable=log_password).grid(row=2,column=1)
    
    Button(log_screen,text='Login',command=Login_session,width=15).grid(row=4,column=1)
    Button(log_screen,text='Close',command=log_screen.destroy,width=15).grid(row=5,column=1)
                             
    global log_notif
    log_notif=Label(log_screen)
    log_notif.grid(row=6,column=1)
##################################################################################################
Label(master,text='Welcome to Bank Application',font=('calibri',20,'bold'),fg='black',bg='green').grid(row=0,column=0)
Button(master,text='Register',command=Register,width=30).grid(row=1,column=0)
Button(master,text='Login',command=Login,width=30).grid(row=3,column=0)
master.mainloop()
```
